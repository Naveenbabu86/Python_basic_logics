```python
"""
Question: Find the Second Largest Number in a List (Without Using max() Twice)

Write a Python function that returns the second largest number from a given list without sorting and without using max() more than once.

Input: [10, 20, 4, 45, 99]
Output: 45

"""
# Input = [10, 20, 4, 45, 99]
# soretedlist = sorted(Input)
# print(soretedlist)
# soretedlist[-2]

numbers = [10, 20, 4, 45, 99]

first = second = float('-inf')  # start with the smallest possible value

for num in numbers:
    if num > first:
        # current number is the new largest
        second = first
        first = num
    elif num > second and num != first:
        # current number is between first and second
        second = num

print("Second largest number is:", second)



```
