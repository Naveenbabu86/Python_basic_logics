```python
s = "abcbbcbb"
seen = set()
start = 0
max_length = 0

for end in range(len(s)):
    while s[end] in seen:
        seen.remove(s[start])
        start += 1
        print("seen value",seen)
        print("start value",start)
    seen.add(s[end])
    print("seen value after adding to set",seen)
    print("end value",end)
    max_length = max(max_length, end - start + 1)

print(max_length)
```

```python
s = "abcbbcbb"
seen = set()
start = 0
max_length = 0
longest_substring = ""  # ðŸ‘ˆ to store the substring itself

for end in range(len(s)):
    while s[end] in seen:
        seen.remove(s[start])
        start += 1

    seen.add(s[end])
    current_substring = s[start:end+1]
    print(current_substring)

    # Check if current substring is the longest
    if len(current_substring) > max_length:
        max_length = len(current_substring)
        longest_substring = current_substring  # ðŸ‘ˆ store it

print("Longest substring without repeating characters:", longest_substring)
print("Length:", max_length)
```
