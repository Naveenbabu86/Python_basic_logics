```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1] * n  # Step 1: Initialize result array with 1s

    # Step 2: Calculate prefix products
    prefix = 1
    for i in range(n):
        answer[i] = prefix
        prefix *= nums[i]   # update prefix product

    # Step 3: Calculate suffix products and multiply
    suffix = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= nums[i]   # update suffix product

    return answer

print(productExceptSelf([1,2,3,4]))



```
