```python
n = [4, 7, 2, 9, 5]

# Initialize two variables with infinity
first = second = float('inf')

for num in n:
    if num < first:
        second = first
        first = num
    elif first < num < second:
        second = num
```

```python
# Find all pairs in a list that sum up to a target number.
# Example:
# Input → [2, 4, 3, 5, 7, 8, 9], Target → 10
# Output → (2, 8), (3, 7), (4, 6)
```

```python
n = [2, 4, 3, 5, 7, 8, 9]
```

```python
nums = [2, 4, 3, 5, 7, 8, 9]
target = 10

seen = set()     # store numbers we've seen
output = set()   # store unique pairs

for num in nums:
    complement = target - num
    if complement in seen:
        # store as a tuple with smaller number first to avoid duplicates
        pair = tuple(sorted((num, complement)))
        output.add(pair)
    seen.add(num)

# convert set to list and print
print(list(output))

```

```python
nums = [2, 4, 3, 5, 7, 8, 9, 5]  # added a second 5
target = 10
output = []

for i in range(len(nums)):
    for j in range(i, len(nums)):  # start j from i, not i+1
        if nums[i] + nums[j] == target:
            pair = tuple(sorted((nums[i], nums[j])))
            if pair not in output:
                output.append(pair)

print(output)
```

```python
n = [1,3,4,2,6,7,3,8,4,9]
target = 10
seen = set()
output = set()


for i in n:
    diff = target - i
    if diff in seen:
        result = tuple(sorted((i,diff)))
        output.add(result)
    seen.add(i)
print(output)


```
