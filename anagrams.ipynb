```python
"""
You are given a list of words, and you need to find groups of anagrams in it.
Words are anagrams if they contain the same letters in a different order.

Expected Output: [['listen', 'silent', 'enlist'], ['rat', 'tar', 'art'], ['cat', 'act']]

"""

words = ["listen", "silent", "enlist", "rat", "tar", "art", "cat", "act"]

anagrams = {}

for word in words:
    # Convert sorted list of characters to tuple for hashable key
    key = tuple(sorted(word))

    if key in anagrams:
        anagrams[key].append(word)
    else:
        anagrams[key] = [word]

# Convert dict values to list of lists
result = list(anagrams.values())

print(result)






```
