```python
"""
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

Expected Output:

[
    [7, 4, 1],
    [8, 5, 2],
    [9, 6, 3]
]

"""

matrix = [[1, 2, 3],[4, 5, 6],[7, 8, 9]]

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Step 1: Transpose (convert rows â†’ columns)
for i in range(len(matrix)):
    for j in range(i, len(matrix)):
        matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

# Step 2: Reverse each row
for row in matrix:
    row.reverse()

print(matrix)

```
