```python
num =  [100, 4, 200, 1, 3, 2]

num = sorted(num)
print(num)
l = []
for i in range(len(num)-1):
    value = num[i+1] - num[i]
    if value == 1:
        l.append(num[i])
print(len(l))

```

```python
num = [100, 4, 200, 1, 3, 2,2,5,6,7,10]
num = sorted(num)

longest = 1
current = 1

for i in range(1, len(num)):
    if num[i] == num[i - 1] + 1:  # consecutive
        current += 1
        longest = max(longest, current)
    elif num[i] != num[i - 1]:    # reset only if not duplicate
        current = 1

print(longest)
```
