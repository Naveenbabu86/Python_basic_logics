```python
matrix = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 8, 9]
]

# Step 1: Find which rows and columns need to be zeroed
rows = set()
cols = set()

for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        if matrix[i][j] == 0:
            rows.add(i)
            cols.add(j)
print(rows)
print(cols)
# Step 2: Set entire row or column to zero
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        if i in rows or j in cols:
            matrix[i][j] = 0

# Final Result
for row in matrix:
    print(row)
```
