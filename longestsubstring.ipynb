```python
"""
Question 5: Find the longest substring without repeating characters

Given a string, find the length of the longest substring that has no repeating characters.

"""
```

```python
s = "abcabcbb"

"""
We maintain a window of unique characters using two pointers:

start: where the current substring begins

end: where the current substring ends

The seen set keeps track of unique characters in the current window.
"""


start = 0
max_len = 0
seen = set()

for end in range(len(s)):
    while s[end] in seen:
        seen.remove(s[start])
        print(seen)
        start += 1
    seen.add(s[end])
    print(seen)
    max_len = max(max_len, end - start + 1)
print(max_len)
```
